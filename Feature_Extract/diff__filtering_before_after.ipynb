# 데이터셋 구성

```python
data=pd.read_csv("/content/drive/MyDrive/weka/dataset/20221115_Person_09_rec_1.csv", skiprows=[1,2])
```

```python
data1=data['signal'].str.strip("[")
data2=data1.str.strip("]")
data3 = data2.str.split(',', expand=True)
```

```python
data_num=data['row_num']
```

\

```python
dataset=pd.concat([data_num, data3], axis=1)
```

```python
dataset.columns = ['row_num', 'before', 'after']
```

```python
dataset
```

```python
dataset["after"]=dataset["after"].astype('float')
dataset["before"]=dataset["before"].astype('float')
```

```python
dataset.info()
```

# 필터링 전후 차이에 대한 통계값

```python
# 각 주기의 max 값 추출

a=20
b=420
list_a = []

for i in range(0,25):
  list_a.append(dataset['after'].iloc[a:b].max())
  a+=400
  b+=400
```

```python
# 각 주기의 min 값 추출

a=20
b=420
list_b = []

for i in range(0,25):
  list_b.append(dataset['after'].iloc[a:b].min())
  a+=400
  b+=400
```

```python
dataset_minmax=pd.DataFrame()

dataset_minmax['max']=pd.DataFrame(list_a)
dataset_minmax['min']=pd.DataFrame(list_b)
```

```python
 dataset_minmax
```

```python
# 각 주기의 max-min 값 추출

datadiff = dataset_minmax.diff(axis=1)
datadiff = datadiff.drop(['max'], axis=1)
```

```python
dataset=pd.concat([dataset_minmax, datadiff], axis=1)
dataset.columns = ['max', 'min', 'diff']
```
