# **모든 데이터에 대한 구간 추출**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
def Data_Preprocessing_oneCycle(file_name, num):
  file_full = dataset_path + '/' + file_name
  df = pd.read_csv(file_full, skiprows=[1, 2])
  
  stripdata=df['signal'].str.strip("[")
  stripdata=stripdata.str.strip("]")
  data = stripdata.str.split(',', expand=True)

  data_num=df['row_num']
  dataset=pd.concat([data_num, data], axis=1)

  
  dataset.columns = ['row_num', 'before', 'after']

  dataset["after"]=dataset["after"].astype('float')
  dataset["before"]=dataset["before"].astype('float')

  sp = round(10000/num)+20
  datasets = dataset[20:sp]
  sns.lineplot("row_num", "after", data=datasets)

  plt.ylabel("after")
  plt.xlabel("timestamp")
  plt.gcf().set_size_inches(50, 15)

  plt.show()
  print(file_name[-12:-10] + "번 사람 데이터 시각화")
  print("- 25초 동안 총", num, "번의 주기를 가짐")

  return 0
```
